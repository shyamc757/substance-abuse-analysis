```python
import pandas as pd
import numpy as np
```

```python
df_2004 = pd.read_csv('DAWN_Data_2004.csv')
df_2005 = pd.read_csv('DAWN_Data_2005.csv')
df_2006 = pd.read_csv('DAWN_Data_2006.csv')
df_2007 = pd.read_csv('DAWN_Data_2007.csv')
df_2008 = pd.read_csv('DAWN_Data_2008.csv')
df_2009 = pd.read_csv('DAWN_Data_2009.csv')
df_2010 = pd.read_csv('DAWN_Data_2010.csv')
df_2011 = pd.read_csv('DAWN_Data_2011.csv')
```

```python
# To check the unique values of each column
def uniques(df):
    
    for column in df.columns:
        
        print(f'\nProcessing Column is: {column}')
        
        # printing unique values in each column to see the cleanliness of the data
        unique_vals = df[column].unique() 
        print(f'\nThe unique values in column {column} are \n{unique_vals}')
    return

def FrequencyOfValues(df):
    
    for column in df.columns:
        
        print(f'\nProcessing Column is: {column}')
        
        # printing frequency of each values in column to check how many unwanted values are there
        count_values = df[column].value_counts()
        print(f'\nThe frequency of each value in column {column} is \n{count_values}')
    return

def mapping(df):
    
    # based on the codebook changing the integer codes to their corresponding values
    
    metro_mapping = {1:'MA',2:'NY',3:'IL',4:'MI',5:'MN',6:'FL',7:'TX',8:'LA',9:'CO',10:'AZ',11:'CA-SD',12:'CA-SF',13:'WA',14:'Other'}
    agecat_mapping = {1:'<=5',2:'6-11',3:'12-17',4:'18-20',5:'21-24',6:'25-29',7:'30-34',8:'35-44',9:'45-54',10:'55-64',11:'>=65',-8:'PNS'}
    sex_mapping = {1:'M',2:'F',-8:'PNS'}
    race_mapping = {1:'white',2:'black or african american',3:'hispanic or latino',4:'all other races',-8:'PNS'}
    casetype_mapping = {1:'suicide attempt',2:'seeking detox',3:'alcohol only(age < 21)',4:'adverse reaction',5:'overmedication',6:'malicious poisoning',7:'accidental ingestion',8:'other'}
    disposition_mapping = {1:'discharged home',2:'released to police/jail',3:'reffered to detox/treatment',4:'ICU/critical care',5:'surgery',6:'chemical dependency/detox, psychiatric unit',7:'other inpatient unit',8:'transferred',9:'left against medical advice',10:'died',96:'other',-8:'PNS'}
    
    df['METRO'] = df['METRO'].replace(metro_mapping)
    df['AGECAT'] = df['AGECAT'].replace(agecat_mapping)
    df['SEX'] = df['SEX'].replace(sex_mapping)
    df['RACE'] = df['RACE'].replace(race_mapping)
    df['CASETYPE'] = df['CASETYPE'].replace(casetype_mapping)
    df['DISPOSITION'] = df['DISPOSITION'].replace(disposition_mapping)
    
    return
```
