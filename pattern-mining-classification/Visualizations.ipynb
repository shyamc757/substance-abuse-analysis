```python
import seaborn as sns


sns.set(style="whitegrid")


plt.figure(figsize=(10, 6))

ax = sns.barplot(x=df_2004_2011['METRO'].value_counts().values,
                 y=df_2004_2011['METRO'].value_counts().index,
                 palette="viridis")

plt.title('Number of Drug Abuse Cases in Different States', fontsize=16)


plt.xlabel('Number of Cases', fontsize=14)
plt.ylabel('States of US', fontsize=14)


for p in ax.patches:
    ax.annotate(f'{p.get_width():,.0f}', (p.get_width() + 0.5, p.get_y() + p.get_height() / 2),
                ha='left', va='center', fontsize=10, color='black')

plt.show()
```

```python

categories = ['<=5', '6-11', '12-17', '18-20', '21-24', '25-29', '30-34', '35-44', '45-54', '55-64', '>=65']

A = list()
B = list()
C = list()
D = list()

grouped_df_2004_2011 = df_2004_2011.groupby(['AGECAT', 'RACE']).size().reset_index(name = 'Number_of_Cases')
grouped_df_2004_2011_sorted = grouped_df_2004_2011.sort_values(by='AGECAT', key=lambda x: x.map({k: i for i, k in enumerate(categories)}))

num_categories = len(categories)

angles = np.linspace(0, 2 * np.pi, num_categories, endpoint=False).tolist()

for i in range(len(grouped_df_2004_2011_sorted)):
    
    if(grouped_df_2004_2011_sorted.iloc[i,1] == 'white' and grouped_df_2004_2011_sorted.iloc[i,0] != 'PNS'): 
        A.append(grouped_df_2004_2011_sorted.iloc[i,2])
    if(grouped_df_2004_2011_sorted.iloc[i,1] == 'black or african american' and grouped_df_2004_2011_sorted.iloc[i,0] != 'PNS'): 
        B.append(grouped_df_2004_2011_sorted.iloc[i,2])
    if(grouped_df_2004_2011_sorted.iloc[i,1] == 'hispanic or latino' and grouped_df_2004_2011_sorted.iloc[i,0] != 'PNS'): 
        C.append(grouped_df_2004_2011_sorted.iloc[i,2])
    if(grouped_df_2004_2011_sorted.iloc[i,1] == 'all other races' and grouped_df_2004_2011_sorted.iloc[i,0] != 'PNS'): 
        D.append(grouped_df_2004_2011_sorted.iloc[i,2])

plt.figure(figsize=(7, 7))
plt.polar(angles, A, label='White', linestyle='solid', linewidth=2, marker='o')

plt.polar(angles, B, label='Black/African American', linestyle='solid', linewidth=2, marker='o')

plt.polar(angles, C, label='Hispanic/Latino', linestyle='solid', linewidth=2, marker='o')

plt.polar(angles, D, label='Other Races', linestyle='solid', linewidth=2, marker='o')

plt.fill(angles, A, alpha=0.25)
plt.fill(angles, B, alpha=0.25)
plt.fill(angles, C, alpha=0.25)
plt.fill(angles, D, alpha=0.25)

plt.xticks(angles, categories)

plt.legend(loc='upper right')

plt.show()
```

```python
# spider plot of number of cases of different sex in different states
categories = ['AZ','CA-SD','CA-SF','CO','FL','IL','LA','MA','MI','MN','NY','TX','WA','Other']

A = list()
B = list()

grouped_df_2004_2011 = df_2004_2011.groupby(['METRO', 'SEX']).size().reset_index(name = 'Number_of_Cases')
grouped_df_2004_2011_sorted = grouped_df_2004_2011.sort_values(by='METRO', key=lambda x: x.map({k: i for i, k in enumerate(categories)}))

num_categories = len(categories)

angles = np.linspace(0, 2 * np.pi, num_categories, endpoint=False).tolist()

for i in range(len(grouped_df_2004_2011_sorted)):
    
    if(grouped_df_2004_2011_sorted.iloc[i,1] == 'M'): 
        A.append(grouped_df_2004_2011_sorted.iloc[i,2])
    if(grouped_df_2004_2011_sorted.iloc[i,1] == 'F'): 
        B.append(grouped_df_2004_2011_sorted.iloc[i,2])

plt.figure(figsize=(7, 7))
plt.polar(angles, A, label='Male', linestyle='solid', linewidth=2, marker='o')

plt.polar(angles, B, label='Female', linestyle='solid', linewidth=2, marker='o')

plt.fill(angles, A, alpha=0.25)
plt.fill(angles, B, alpha=0.25)

plt.xticks(angles, categories)

plt.legend(loc='upper right')

plt.show()
```
